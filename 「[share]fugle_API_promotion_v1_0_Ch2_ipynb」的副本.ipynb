### 使用行情快照（snapshot）取得熱門個股

取當日成交量上市前 5 大的股票 <br>
單量超過 20 張，就進行 line notify 通知

```python
# # 取當日成交量上市前 5 大的股票列表
symbol_list = list(map(lambda x:x['symbol'], stock.snapshot.actives(market='TSE', trade='volume')['data'][:5]))

# 訂定單量
size_limit = 20
```

```python
class StockNotifier:
    def __init__(self, api_key, line_token, symbol_list, size_limit):
        self.api_key = api_key
        self.line_token = line_token
        self.symbol_list = symbol_list
        self.size_limit = size_limit
        self.client = None

    def _on_new_price(self, message):

        print(message)
        json_data = json.loads(message)

        if json_data['event']=="data" and json_data.get('data',{}).get('isTrial') == None: # 避免用到試撮資料

            # 取最新成交價
            latest_price = json_data['data']['price']
            # 取最新單量
            size = json_data['data']['size']
            # 取最新總量
            volume = json_data['data']['volume']

            symbol = json_data['data']['symbol']

            # 如果最新單量超過訂定的單量張數，就進行 line notify 通知！
            if size > self.size_limit:

                # 預計回傳 line notify 通知的訊息
                msg = f'\n{symbol} 單量超過 {str(self.size_limit)} 張！\n最新價格 {str(latest_price)} 元\n單量 {str(size)} 張\n總量 {str(volume)} 張\n快到 https://www.fugle.tw/ai/{symbol} 查查！ '

                # 執行 line notify 通知
                lineTool.lineNotify(self.line_token, msg)

        # # 盤後測試可以加上以下判斷
        elif json_data['event']=="snapshot":

            # 取最新成交價
            latest_price = json_data['data']['price']
            # 取最新單量
            size = json_data['data']['size']
            # 取最新總量
            volume = json_data['data']['volume']

            symbol = json_data['data']['symbol']

            # 如果最新單量超過訂定的單量張數，就進行 line notify 通知！
            if size > self.size_limit:

                # 預計回傳 line notify 通知的訊息
                msg = f'\n{symbol} 單量超過 {str(self.size_limit)} 張！\n最新價格 {str(latest_price)} 元\n單量 {str(size)} 張\n總量 {str(volume)} 張\n快到 https://www.fugle.tw/ai/{symbol} 查查！ '

                # 執行 line notify 通知
                lineTool.lineNotify(self.line_token, msg)


    def handle_connect(self):
        print('connected')

    def handle_disconnect(self, code, message):
        print(f'disconnect: {code}, {message}')

    def handle_error(self, error):
        print(f'error: {error}')

    def main(self):
        self.client = WebSocketClient(api_key=self.api_key)
        stock = self.client.stock
        stock.on('message', self._on_new_price)
        stock.on("connect", self.handle_connect)
        stock.on("disconnect", self.handle_disconnect)
        stock.on("error", self.handle_error)

        stock.connect()

        stock.subscribe({
            'channel': 'trades',
            'symbols': self.symbol_list
        })

if __name__ == '__main__':
    notifier = StockNotifier(api_key=fugle_api_key, line_token=line_token, symbol_list=symbol_list, size_limit=size_limit)
    notifier.main()
```

## 補充教材 - 取當日成交量上市前 5 大的股票清單，當即時報價突破 5日線時，進行 line notify 通知

```python
## 取當日成交量上市前 5 大的股票列表
symbol_list = list(map(lambda x:x['symbol'], stock.snapshot.actives(market='TSE', trade='volume',type="COMMONSTOCK")['data'][:5]))
symbol_list
```

```python
import json
from fugle_marketdata import RestClient, WebSocketClient
import lineTool

class StockTechnicalNotifier:
    def __init__(self, api_key, line_token, symbol_list):
        self.api_key = api_key
        self.line_token = line_token
        self.symbol_list = symbol_list
        self.client = None
        self.fugle_client = RestClient(api_key=self.api_key)
        self.sma_cache = self.get_all_sma_5()  # 初始化時一次性獲取所有股票的 5MA
        self.previous_prices = {symbol: None for symbol in symbol_list}  # 記錄上一筆成交價

    def get_sma_5(self, symbol):
        """使用 Fugle SDK 取得指定股票的 5 日均線"""
        try:
            response = self.fugle_client.stock.technical.sma(**{
                "symbol": symbol,
                "from": "2024-03-01",  # 取最近的數據範圍
                "to": "2024-03-10",
                "timeframe": "D",
                "period": 5
            })

            if response and 'data' in response and response['data']:
                return round(response['data'][-1]['sma'],3) # 取得最新的 5MA 數值

        except Exception as e:
            print(f"取得 {symbol} 5MA 失敗: {e}")

        return None  # 若 API 請求失敗，回傳 None

    def get_all_sma_5(self):
        """一次性獲取所有監測股票的 5 日均線，並存入快取"""
        sma_data = {}
        for symbol in self.symbol_list:
            sma_data[symbol] = self.get_sma_5(symbol)
        print(f"初始化 5MA 數據: {sma_data}")  # 確保數據正確
        return sma_data

    def _on_new_price(self, message):
        json_data = json.loads(message)
        print(json_data)

        if json_data['event'] == "data" and json_data.get('data', {}).get('isTrial') is None:  # 避免用到試撮資料
            latest_price = json_data['data']['price']
            symbol = json_data['data']['symbol']

            # **使用初始化時的 5MA 值進行比較**
            sma_5 = self.sma_cache.get(symbol)
            previous_price = self.previous_prices.get(symbol)  # 取得上一筆價格

            # **檢查是否發生突破**
            if sma_5 and previous_price is not None:
                if previous_price <= sma_5 and latest_price > sma_5:
                    msg = (f'\n{symbol} **股價正式突破 5 日均線！**'
                           f'\n最新價格 {latest_price} 元\n5MA 均線 {sma_5} 元'
                           f'\n快到 https://www.fugle.tw/ai/{symbol} 查查！')
                    lineTool.lineNotify(self.line_token, msg)

            # **更新上一筆價格**
            self.previous_prices[symbol] = latest_price

    def handle_connect(self):
        print('connected')

    def handle_disconnect(self, code, message):
        print(f'disconnect: {code}, {message}')

    def handle_error(self, error):
        print(f'error: {error}')

    def main(self):
        self.client = WebSocketClient(api_key=self.api_key)
        stock = self.client.stock
        stock.on('message', self._on_new_price)
        stock.on("connect", self.handle_connect)
        stock.on("disconnect", self.handle_disconnect)
        stock.on("error", self.handle_error)

        stock.connect()

        stock.subscribe({
            'channel': 'trades',
            'symbols': self.symbol_list
        })

if __name__ == '__main__':
    notifier = StockTechnicalNotifier(api_key=fugle_api_key, line_token=line_token, symbol_list=symbol_list)
    notifier.main()
```
